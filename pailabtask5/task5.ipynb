```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread(r"C:\Users\dell\Desktop\pailabtask5\catoo.jpeg.crdownload")

if image is None:
    raise FileNotFoundError("Image not found or unable to read the image file.")

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# 1
resized = cv2.resize(image_rgb, (300, 300))
# 2
gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
# 3
blurred = cv2.GaussianBlur(image_rgb, (7, 7), 0)
# 4
edges = cv2.Canny(gray, 100, 200)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(gray, cmap='gray')
axs[0].set_title("Grayscale")
axs[1].imshow(blurred)
axs[1].set_title("Blurred")
axs[2].imshow(edges, cmap='gray')
axs[2].set_title("Canny Edges")
for ax in axs: ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
# 2
orb = cv2.ORB_create()
# 3
keypoints, descriptors = orb.detectAndCompute(gray, None)
# 4
feature_img = cv2.drawKeypoints(image_rgb, keypoints, None, color=(0, 255, 0))
```

```python
plt.imshow(feature_img)
plt.title("ORB Feature Keypoints")
plt.axis("off")
plt.show()
```

```python
# 1
canvas = image_rgb.copy()
# 2
cv2.line(canvas, (50, 50), (250, 50), (255, 0, 0), 3)
# 3
cv2.rectangle(canvas, (60, 60), (240, 240), (0, 255, 0), 2)
# 4
cv2.circle(canvas, (150, 150), 40, (0, 0, 255), -1)
# 5
cv2.putText(canvas, "OpenCV Rocks", (50, 280), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 0), 2)
```

```python

plt.imshow(canvas)
plt.title("Drawing Shapes and Text")
plt.axis("off")
plt.show()
```
